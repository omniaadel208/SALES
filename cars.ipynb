```python
# group by year and month and get the sum of SALES
monthly_sales = df.groupby(['Year', 'Month'])['SALES'].sum().reset_index()

# for each year get the month with the max sales
max_sales_per_year = monthly_sales.loc[monthly_sales.groupby('Year')['SALES'].idxmax()].reset_index(drop=True)

# sort by year
max_sales_per_year.sort_values('Year', inplace=True)
```

```python
df_country = pd.crosstab(df['COUNTRY'], df['PRODUCTLINE'], values=df['QUANTITYORDERED'], aggfunc='sum')
```

### plotting the number of quantity ordered in each product line in each country when stacked = true

```python
df_country.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Stacked Bar Chart: Quantity Ordered by Country and Product Line')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_country, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Quantity Ordered by Country and Product Line')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### plotting the number of quantity ordered in each product line in each country when stacked = false

```python
df_country.plot(kind='bar', stacked=False, figsize=(12, 7))
plt.title('Stacked Bar Chart: Quantity Ordered by Country and Product Line')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Compute percentage per row
df_country_pct = df_country.div(df_country.sum(axis=1), axis=0) * 100
ax = df_country_pct.plot(kind='bar', stacked=True, figsize=(12, 7), cmap='tab20')
plt.title('Stacked Bar Chart (Percentages): Quantity Ordered by Country and Product Line')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.legend(title='Product Line', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
for c in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 5 else '' for v in c]  # Hide small ones
    ax.bar_label(c, labels=labels, label_type='center', fontsize=8, color='black')

plt.show()
```
